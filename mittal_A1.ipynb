## **Part 1**: Data Science Salaries

#### 1.1: Read the CSV

Load the salaries.csv into a DataFrame

```python
salaries = pd.read_csv('salaries.csv') #reading the file using the read_csv function
salaries.head(20)
```

#### 1.2: Exchange Rates

In the cell below, use the exchange rate below to convert the USD salaries to CAD. **Then**, drop the USD column. Save into a new DataFrame called "Updated_Salaries", which will be used for the following questions.

Exchange rate: 0.69502999 US Dollars for 1 CAD.  

Credit: https://www.xe.com/currencyconverter/convert/?Amount=1&From=CAD&To=USD

```python
salaries['salary_in_CAD'] = np.round(salaries['salary_in_usd']/0.69502999,2) # fetching the column, then converting and rounding to make it equal to CAD and then fropping the USD salary
updated_salaries = salaries.drop('salary_in_usd', axis=1)
updated_salaries.head(20)
```

Write a code to calculate and display the average salary of Entry-Level data science employees. Group the data by company size and year. Present the output as a table.

```python
# Filtering the data for Entry-Level data science roles
entry_level = updated_salaries[updated_salaries['experience_level'] == 'Entry-Level']

# Calculating the average (using mean function) salary grouped by company size and year
average_salary_by_size_year = entry_level.groupby(['company_size', 'work_year'])['salary_in_CAD'].mean().reset_index()

average_salary_by_size_year
```

#### 2.2: Frequency Distribution of Numerical Quantity.



Show the salary distribution in the year of 2021 and 2023.

**Note**: You have to show **two separate distributions in one graph**, using **different colors**. Use 20 equal-sized bins ranging from the maximum to the minimum value in your distribution.

###2.2.1: Frequency Distribution  

In the cell below, create a Frequency Distribution of salary.

```python
# Filtering data for the years 2021 and 2023
salaries_2021 = updated_salaries[updated_salaries['work_year'] == 2021]['salary_in_CAD']
salaries_2023 = updated_salaries[updated_salaries['work_year'] == 2023]['salary_in_CAD']

# Calculating the range for bins
min_salary = min(salaries_2021.min(), salaries_2023.min())
max_salary = max(salaries_2021.max(), salaries_2023.max())
bins = np.linspace(min_salary, max_salary, 20)

# Plotting the required distribution
plt.figure(figsize=(12, 6))
plt.hist(salaries_2021, bins=bins, alpha=0.7, label='2021', color='black')
plt.hist(salaries_2023, bins=bins, alpha=0.7, label='2023', color='chocolate')
plt.xlabel('Salary in CAD')
plt.ylabel('Number of Employees')
plt.title('Salary Distribution in 2021 vs 2023')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###2.2.2: Relative Frequency Distribution


In the cell below, create a Relative Frequency Distribution of Salary.

```python
# Created a histogram with relative frequency using the same bins as before
plt.figure(figsize=(12, 6))
plt.hist(updated_salaries['salary_in_CAD'], bins=bins, weights=np.ones(len(updated_salaries['salary_in_CAD'])) / len(updated_salaries['salary_in_CAD']), color='purple')
plt.xlabel('Salary in CAD')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Distribution of Salary_in_CAD')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## **Part 3**: Visualizing Trends
**Using the Updated_Salaries Dataframe.**

#### 3.1: Salary Trend

Represent the trend of the **mean salary** **over the years** **for each company size** using a bar chart (Hint: The x-axis can either be the work years or the company sizes). To receive **full marks**, your figure should specify the following:

  
i. Axis labels  
ii. Title

```python
# Grouped by work year and company size, then calculated the mean salary
mean_salaries_by_year_size = updated_salaries.groupby(['work_year', 'company_size'])['salary_in_CAD'].mean().unstack()
colors = ['red', 'blue', 'green']
# Plotting the mean salary trends over the years for each company size using a bar chart
mean_salaries_by_year_size.plot(kind='bar', figsize=(12, 6), color = colors, alpha = 0.8)
plt.xlabel('Work Year')
plt.ylabel('Mean Salary in CAD')
plt.title('Mean Salary Trends Over the Years by Company Size')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Company Size')
plt.show()
```

#### 3.2: Experience vs Salary  

In the cell below, create and display a scatterplot of **experience level** (x-axis) and **average salary** (y-axis), of those that work in **large-sized** companies.



```python
# Filtering data for large-sized companies
large_companies = updated_salaries[updated_salaries['company_size'] == 'L']

# Grouped by experience level and calculate the mean salary
mean_salary_by_experience = large_companies.groupby('experience_level')['salary_in_CAD'].mean().reset_index()

# Sorting the experience levels in a meaningful order
experience_order = {'Entry-Level': 0, 'Mid-Level': 1, 'Senior-Level': 2, 'Junior-Level': 3}
mean_salary_by_experience['experience_order'] = mean_salary_by_experience['experience_level'].map(experience_order)
mean_salary_by_experience = mean_salary_by_experience.sort_values('experience_order')

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(mean_salary_by_experience['experience_level'], mean_salary_by_experience['salary_in_CAD'], color='red')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in CAD')
plt.title('Average Salary by Experience Level in Large-Sized Companies')
plt.grid(True)
plt.show()
```

#### 3.3: Size Matters  

Create another scatterplot based on your scatterplot from 3.2. This time, overlay three scatterplots, each representing a different company size.

**Ensure that the data points are color-coded based on their respective company sizes**. You may use any color scheme, as long as the colors are clearly distinguishable.

For guidance, refer to this:
https://python-graph-gallery.com/537-scatter-plots-grouped-by-color-with-pandas/

```python
# Filtering data for each company size and calculate the mean salary
small_companies = updated_salaries[updated_salaries['company_size'] == 'S']
medium_companies = updated_salaries[updated_salaries['company_size'] == 'M']
large_companies = updated_salaries[updated_salaries['company_size'] == 'L']

# Grouping by experience level and calculate the mean salary for each company size
mean_salary_small = small_companies.groupby('experience_level')['salary_in_CAD'].mean().reset_index()
mean_salary_medium = medium_companies.groupby('experience_level')['salary_in_CAD'].mean().reset_index()
mean_salary_large = large_companies.groupby('experience_level')['salary_in_CAD'].mean().reset_index()

# Sorting by experience level with the defined order for plotting
mean_salary_small['experience_order'] = mean_salary_small['experience_level'].map(experience_order)
mean_salary_medium['experience_order'] = mean_salary_medium['experience_level'].map(experience_order)
mean_salary_large['experience_order'] = mean_salary_large['experience_level'].map(experience_order)

mean_salary_small = mean_salary_small.sort_values('experience_order')
mean_salary_medium = mean_salary_medium.sort_values('experience_order')
mean_salary_large = mean_salary_large.sort_values('experience_order')

# Plotting the scatter plots for each company size
plt.figure(figsize=(10, 6))
plt.scatter(mean_salary_small['experience_level'], mean_salary_small['salary_in_CAD'], color='blue', label='Small Companies')
plt.scatter(mean_salary_medium['experience_level'], mean_salary_medium['salary_in_CAD'], color='green', label='Medium Companies')
plt.scatter(mean_salary_large['experience_level'], mean_salary_large['salary_in_CAD'], color='red', label='Large Companies')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in CAD')
plt.title('Comparison of Average Salary by Experience Level Across Company Sizes')
plt.legend(title='Company Size')
plt.grid(True)
plt.show()
```
