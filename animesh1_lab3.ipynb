### **Part 1**:

The file "universities.csv" contains a ranking of the world's top universities, in descending order.

Credit: https://www.timeshighereducation.com/world-university-rankings

#### Hint: Python [lambda functions](https://sparkbyexamples.com/pandas/pandas-apply-with-lambda-examples/) can greatly streamline simple transformations.

##### 1.1: Read the csv into a DataFrame

```python
data = pd.read_csv('universities.csv')
data.head(5)
```

##### 1.2: Modify the DataFrame such that the `students` column contains integers instead of strings.

```python
data["students"] = data["students"].apply(lambda x: int(x.replace(",", "")))
data.head(5)
```

##### 1.3: Create a `proportion_international` column that includes the proportion of students that are international, represented as a float between 0 and 1.

```python
data["proportion_international"] = data["pct_international"].apply(lambda x: float(x.replace("%", ""))/100)
data.head(5)
```

##### 1.4: Add columns `percent_female` and `percent_male` to the DataFrame. Each column should consist of int values and be in the range of [0,100].

```python
data["percent_female"] = data["f_m_ratio"].apply(lambda x: x[:x.find(":")])
data["percent_male"] = data["f_m_ratio"].apply(lambda x: x[x.find(":")+1:])

data.head(5)
```

##### 1.5: Create the column `num_domestic`. The column should contain ints, representing the number of domestic (non-international) students, rounded to the nearest integer value (Hint: You may find np.round() helpful).

```python
data["num_domestic"] = data.apply(lambda row: int(np.round((1-row["proportion_international"]) * row["students"], 0)), axis=1)
data.head(5)
```

### **Part 2**:

#### 2.1: Let's continue to use our university DataFrame. Remove the columns "pct_international" and "f_m_ratio". Then, display the first 5 rows.

```python
data = data.drop(["pct_international", "f_m_ratio"], axis=1)
data.head(5)
```

#### 2.2: Create and display a DataFrame assigned to the variable `numerical`, containing only the numerical columns. Feel free to do this however you wish.

```python
numerical = data.loc[:, "students":]
numerical.head(5)
```

#### 2.3: Create and display a DataFrame assigned to the variable `categorical`, containing only the categorical columns. Feel free to do this however you wish.

```python
categorical = data.iloc[:, :2]
categorical.head()
```

#### 2.4: Create and display a DataFrame assigned to the variable `most_students`. The DataFrame should contain the 10 countries with the most total students. Include the total number of domestic and international students for each country, as you will need this for 2.5.

```python
data["international_students"] = data['students'] - data['num_domestic']

most_students = data.groupby('location').agg({
    'students': 'sum',
    'num_domestic': 'sum',
    'international_students': 'sum'
}).sort_values(by='students', ascending=False).head(10)

most_students.head(10)
```

#### 2.5: Use your DataFrame from 2.4 to create a stacked bar chart that displays the total number of international and domestic students in each country.

```python
bar_chart = px.bar(most_students, x=most_students.index, y=['num_domestic', 'international_students'], barmode='stack')
bar_chart.show()
```

#### 2.6: Create a bar chart that displays the total number of top 100 schools in each country. Countries without any top 100 school do not need to be included (Hint: The original DataFrame is in descending order for ranking).

```python
top_100_schools = data.iloc[:100]

schools_count = top_100_schools.groupby('location').size().reset_index(name='number_of_schools')

schools_count = schools_count.sort_values('number_of_schools', ascending=False)

fig = px.bar(schools_count, x='location', y='number_of_schools', labels={'location': 'Country', 'number_of_schools': 'Number of Schools'}, orientation='v')

fig.show()
```

#### 2.7: Use your DataFrame from 2.4 to create a pie chart dislaying the proportion of total students relative to the other countries in the DataFrame.

```python
fig = px.pie(
    most_students,
    names=most_students.index,
    values="students",
    height=500,
    width=500
)

fig.show()
```
